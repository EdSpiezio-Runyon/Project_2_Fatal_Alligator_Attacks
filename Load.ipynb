# SQL Load Project 2

```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store and Clean API_weather.csv into DataFrame

```python
csv_file = "Resource/API_weather.csv"
weather_df = pd.read_csv(csv_file)
weather_df.columns = [x.replace(' ', '_').replace('(', '').replace(')', '').lower() for x in weather_df.columns]
weather_df.head()
```

### Store and Clean fatal_alligator_attacks_US_with_latlng.csv into Dataframe

```python
alligator_file = "Resource/fatal_alligator_attacks_US_with_latlng.csv"
alligator_df = pd.read_csv(alligator_file)
alligator_df.columns = [x.replace(' ', '_').lower() for x in alligator_df.columns]
alligator_df = alligator_df.drop(columns=['unix_start_date','unix_end_date'])
alligator_df.head()
```

### Connect to local database

```python
protocol = 'postgresql'
username = 'postgres'
password = 'postgres'
host = 'localhost'
port = 5432
database_name = 'fatal_gator_db'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

### Use pandas to load csv converted DataFrame into database

```python
weather_df.to_sql(name='weather', con=engine, if_exists='append', index=False)
alligator_df.to_sql(name='fatal_gator', con=engine, if_exists='append', index=False)
```

### Confirm data has been added by querying the customer_name table
* NOTE: can also check using pgAdmin

```python
pd.read_sql_query('select * from weather', con=engine).head()
```
